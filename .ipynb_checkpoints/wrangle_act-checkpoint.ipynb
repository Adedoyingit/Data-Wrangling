### Visualization

```python
import matplotlib.pyplot as plt
def plothist(column, x_label, y_label, diagram_title):
    column.hist()
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.title(diagram_title)
    plt.show()
# dfs_clean.favorite_count.hist()
```

```python
plothist(dfs_clean.favorite_count, 'favorite_count', 'frequency', 'Histogram_representing_favorite_count_column')
```

```python
plothist(dfs_clean.img_num, 'img_num', 'frequency', 'Histogram_representing_img_num_column')

#plothist(dfs_clean.value.value_counts(), 'value', 'frequency', 'Histogram_representing_value_column')
```

```python
ax = dfs_clean.value.value_counts().plot(kind = 'bar', title = 'Graph representing the value column')
ax.set_xlabel('value')
ax.set_ylabel('frequency')
```

```python
def plotscatter(x,y, X, Y, diagram_title):
    plt.scatter(x,y)
    plt.xlabel(X)
    plt.ylabel(Y)
    plt.title(diagram_title)
    plt.show()
```

```python
plotscatter(dfs_clean.value, dfs_clean.retweet_count, 'dfs_clean.value', 'dfs_clean.retweet_count', 'Graph showing the relationship between value and retweet_count')
```
